# Seconds execise cleaning data and transform data

```python
import pandas as pd
import numpy as np
```

```python
data = {
    "nombre":  ["  Ana", "Luis ", "  Ana", "Carlos", "  Luis ", "Maria"],
    "edad":    [25, 30, 25, np.nan, 30, 22],
    "ciudad":  ["Madrid ", "  Barcelona", "Madrid ", "Sevilla", "  Barcelona", np.nan],
    "salario": [2500, 3000, 2500, 2800, 3000, np.nan],
}
```

```python
print(data)
```

```python
df = pd.DataFrame(data)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.drop_duplicates()
```

```python
df
```

```python
bins = [0,2000, 5000, float('inf')]
labels = ['Bajo', 'Medio', 'Alto']
df['categoria_salario'] = pd.cut(df['salario'], bins=bins, labels=labels)
```

```python
print(df)
```
